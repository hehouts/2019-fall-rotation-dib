```python
%%bash
#Calc signatures


for org in gingivalis

#bacteroides denticola 
do
    for mol in genomic rna 
    do
        cd $org
        mkdir -p $mol/sigs
        for infile in $mol/seqs/*a.gz 
          do 
            for k in 21 31 51
              do

                out_name=$(basename $infile .fna.gz)
                sourmash compute -k $k --scaled 2000 --track-abundance -o $mol/sigs/${out_name}.sig ${infile}

                #back to relative home dir, overwrite and append this file

             done
          done
        cd ..
    done
    
    
    for mol in protein
    do

        cd $org
        mkdir -p $mol/sigs
        for infile in $mol/seqs/*a.gz 
        do 
        for k in 7 12 15
        do
            out_name=$(basename $infile .faa.gz)
            sourmash compute --input-is-protein -k $k --scaled 2000 --track-abundance -o $mol/sigs/${out_name}.sig ${infile}

        done
        done
        cd ..
    done
   
   
done


```

```python
%%bash


##For dayhoff and HP
#Calc signatures


for org in gingivalis
#bacteroides denticola 
do
    for mol in protein
    do

        cd $org
        mkdir -p $mol/sigs/dayhoff
        mkdir -p $mol/sigs/hp
        
        for infile in $mol/seqs/*a.gz 
        do 
            for k in 7 12 15
            do
                out_name=$(basename $infile .faa.gz)

                #Dayhoff
                sourmash compute --input-is-protein --dayhoff -k $k --scaled 2000 --track-abundance -o $mol/sigs/dayhoff/${out_name}.sig ${infile}

                #HP
                sourmash compute --input-is-protein --hp -k $k --scaled 2000 --track-abundance -o $mol/sigs/hp/${out_name}.sig ${infile}

            done
        done
        cd ..
    done
   
   
done
```
