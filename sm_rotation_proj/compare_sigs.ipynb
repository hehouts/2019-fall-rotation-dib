```python
# Use sourmash compare to compare signatures that were calculated with 
# the same k size, the same molecule type, and the same encoding (e.g. dayoff or h-p).

# Take a look at the metapallette paper for some applications of k-sizes to detection of relatedness across
# evolutionary distances, or take our word that k = 21 ~ genus level similarlity, k = 31 ~ species level 
# similarity, and k = 51 ~ strain level similarity. https://msystems.asm.org/content/1/3/e00020-16
```

```python
%%bash
less calcd_k_amino_acids.txt

less calcd_k_nuclc_acids.txt
```

```python
%%bash


for org in bacteroides
#denticola gingivalis
do
    for mol in genomic 
#rna 
    do
        for k in 21 
#31 51
        do
        #Makes a compare folder for each k size
            mkdir -p $org/$mol/sm_compare_tables
            
            sourmash compare -k $k -o ${org}/${mol}/sm_compare_tables/${org}_k${k}_${mol}_compare  ${org}/${mol}/sigs/*.sig


        #Makes a folder for sm_compare_table
            mkdir -p $org/$mol/sm_plots        
            sourmash plot --output-dir ${org}/${mol}/sm_plots --labels --pdf  ${org}/${mol}/sm_compare_tables/${org}_k${k}_${mol}_compare    

            
        #This is probably computationally redundant    
            mkdir -p $org/$mol/sm_comp_CSVs
            sourmash compare -k $k --csv ${org}/${mol}/sm_comp_CSVs/${org}_k${k}_${mol}_compare.csv $org/$mol/sigs/*.sig
        done

     #once per $org/$mol
     #Send the .csv's and the pdfs to the Windows desktop
            cp $org/$mol/sm_plots/*.pdf /mnt/c/Users/user/Desktop/Sourmash_Outputs/PDFs/
            cp $org/$mol/sm_comp_CSVs/*.csv /mnt/c/Users/user/Desktop/Sourmash_Outputs/CSVs/
            

    done
   
done

```
